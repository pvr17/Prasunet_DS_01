```python
import pandas as pd
import matplotlib.pyplot as plt
#Loading the dataset
df = pd.read_csv("//Users//sahithipadamati//Downloads//T1dataset//API_SP.POP.TOTL_DS2_en_csv_v2_340879.csv", skiprows=3)
#Let's choose Afghanistan as an example
country = 'Afghanistan'
country_data = df[df['Country Name'] == country]
#Extract years and population values
years = country_data.columns[4:]  #assuming columns from 4 onwards are years
population = country_data.iloc[0, 4:].values
#Convert population values to numeric
population = pd.to_numeric(population, errors='coerce')
#Plotting
plt.figure(figsize=(11, 7))
plt.bar(years, population, color='blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {country}')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram ')
plt.show()
```
